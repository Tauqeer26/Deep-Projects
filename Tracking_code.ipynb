```python
#Add 3 points angle

import cv2 
import cv2 as cv
import numpy as np
import math
cap=cv2.VideoCapture("F:/C_new.wmv")

ret,frame = cap.read()
# setup initial location of window
bbox=cv2.selectROI('Tracking',frame,False)
x, y, w, h = int(bbox[0]),int(bbox[1]),int(bbox[2]),int(bbox[3])# simply hardcoded the values
track_window = (x, y, w, h)
# set up the ROI for tracking
roi = frame[y:y+h, x:x+w]
hsv_roi =  cv.cvtColor(roi, cv.COLOR_BGR2HSV)
mask = cv.inRange(hsv_roi, np.array((0., 60.,32.)), np.array((180.,255.,255.)))
roi_hist = cv.calcHist([hsv_roi],[0],mask,[180],[0,180])
cv.normalize(roi_hist,roi_hist,0,255,cv.NORM_MINMAX)
# Setup the termination criteria, either 10 iteration or move by atleast 1 pt
term_crit = ( cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 10, 1 )
while(1):
    ret, frame = cap.read()
    if ret == True:
        hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
        dst = cv.calcBackProject([hsv],[0],roi_hist,[0,180],1)
        
        # apply camshift to get the new location
        ret, track_window = cv.CamShift(dst, track_window, term_crit)
        #cv2.circle(img, (int(track_window[2]),int(track_window[3])), 9, (255,255,0), -1)
        # Draw it on image
        center = (track_window[0]+(track_window[2]/2), track_window[1]+(track_window[3]/2))
        pts = cv.boxPoints(ret)
        
        #print(pts)
        #cv2.circle(frame,(pts[0][1],pts[0][2]))
        pts = np.int0(pts)
        cv2.circle(frame, (630,350), 7, (0,0,255), -1)
        cv2.circle(frame, (630,100), 9, (255,25,0), -1)
        diffx=abs(track_window[0]-630)
        diffy=abs(track_window[1]-350)
        print("This is Y",diffy)
        print("This is X",diffx)
        per=math.atan2(diffy,diffx)
        p=math.degrees(per)
        cv2.putText(frame,str(int(p)),(75,45),cv2.FONT_HERSHEY_COMPLEX,0.7,(255,0,255),2)
        cv2.circle(frame, (int(center[0]),int(center[1])), 7, (255,0,255), -1)
        
        img2 = cv.polylines(frame,[pts],True, 255,2)
        
        #cv2.circle(img2, (), 7, (0,0,255), -1)
        
        cv.imshow('img2',img2)
        if cv2.waitKey(1)& 0xff==ord('q'):
            cv2.destroyAllWindows()
            break
        
```
